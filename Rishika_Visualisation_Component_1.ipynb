##Univariate Plots
We try to plot the univariate graphs of the variables to draw out conclusions. 

```r
colnames(police_data)[colnames(police_data) == 'Date.of.injury.resulting.in.death.DD.MM.YYYY.'] ='Date.of.injury'
```

```r
police_data$year=format(as.Date(police_data$Date.of.injury,"%d-%m-%Y"),"%Y")
unique(police_data$year)
```

```r
ggplot(police_data,aes(x=year,fill=year))+geom_bar()+coord_flip()+ggtitle("Years when the innocent deaths took place")+xlab("Year")+ylab("Frequency")+theme(legend.position="NONE")+scale_fill_viridis_d()
```

```r
sum(police_data$year=="2020")/length(police_data$year)
sum(police_data$year=="2000")/length(police_data$year)
```


*   2020 was the year in which most of the innocent deaths took place with a contribution of about 6.39% of the total deaths in 2 decades.

*  2000 was the year in which the least number of innocent deaths happened with a contribution of about 2.77% of the total deaths in 2 decades.

## Bivariate/Multivariate Plots
